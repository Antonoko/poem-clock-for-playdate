```python
KK_QUOTE_TEXT = 'kk-quote.txt'
KK_QUOTE_JSON = 'kk-quote.json'
```

```python
# 读取配置
import json

def read_json_file(file_name: str):
    with open(file_name, 'r', encoding="utf-8") as file:
        data = json.load(file)
    return data

def write_dict_to_json(file_path, data):
    with open(file_path, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=4)
```

```python
# 导入需要的库
import os

# 初始化一个空的列表存放段落内容
paragraphs = []

# 定义一个函数来读取文件
def read_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.readlines()
    return content

# 定义一个函数来处理文本并将段落保存到列表中
def get_paragraphs(content):
    paragraph = ""
    for line in content:
        if '● • ————' in line:
            if paragraph:
                paragraphs.append(paragraph.strip())
                paragraph = ""
        else:
            paragraph += line
    if paragraph:
        paragraphs.append(paragraph.strip())

# 定义主函数
def main():
    file_path = KK_QUOTE_TEXT
    content = read_file(file_path)
    get_paragraphs(content)

main()

# 输出段落列表
for index, paragraph in enumerate(paragraphs):
    print(f"Paragraph {index + 1}:\n{paragraph}")
```

```python
def merge_lines(string):
    threshold = 12

    lines = string.split('\n')
    merged_lines = []
    buffer_line = ""

    for line in lines:
        buffer_line += line
        if len(buffer_line) >= threshold:
            merged_lines.append(buffer_line)
            buffer_line = ""
            
    # 如果剩余的部分长度仍小于8，将其合并到最后一行
    if buffer_line and len(buffer_line) < threshold and merged_lines:
        merged_lines[-1] += buffer_line
    elif buffer_line:
        merged_lines.append(buffer_line)

    return '\n'.join(merged_lines)

format_paragraphs = []

for quote in paragraphs:
    lines = quote.split('\n')
    if len(lines) > 6:
        quote = merge_lines(quote)
    format_paragraphs.append(quote)
```

```python
a = 0
for quote in format_paragraphs:
    lines = quote.split('\n')
    if len(lines) > 4:
        a+=1
print(a)
```

```python
json_res = {
    "kk_quote": format_paragraphs
}
write_dict_to_json(KK_QUOTE_JSON, json_res)
```
